```python
def main():
    print("Добро пожаловать на кастинг в кино!")
    
    # Запрос о желании сняться в кино
    choice = input("Хотите ли вы сняться в кино? (да/нет): ").strip().lower()
    
    if choice == 'да':
        print("\nОтлично! Предлагаем вам две роли на выбор:")
        print("1. Халк - могучий супергерой")
        print("2. Локи - хитрый бог обмана")
        
        role_choice = input("Выберите роль (Халк/Локи): ").strip().lower()
        
        if role_choice == 'хулк' or role_choice == 'халк':
            # Ветка для Халка
            try:
                bicep_size = float(input("\nСколько у вас бицепс в объеме (в см)? "))
                
                if bicep_size < 60:
                    print("Извините, но ваши бицепсы слишком малы. Отправляйтесь домой качаться!")
                else:
                    print("Поздравляем! Вы приняты на роль Халка!")
                    
            except ValueError:
                print("Ошибка! Пожалуйста, введите число для объема бицепса.")
                
        elif role_choice == 'локи':
            # Ветка для Локи
            parent_choice = input("\nКого вы больше любите - маму или папу? ").strip().lower()
            
            if parent_choice == 'папу' or parent_choice == 'папа':
                print("Извините, но такие предпочтения нам не подходят. Отправляйтесь домой!")
            elif parent_choice == 'маму' or parent_choice == 'мама':
                print("Отправляйтесь к папе, чтобы спросить, кого вы больше любите!")
            else:
                print("Ошибка! Пожалуйста, выберите между мамой и папой.")
                
        else:
            print("Ошибка! Пожалуйста, выберите одну из предложенных ролей: Халк или Локи.")
            
    elif choice == 'нет':
        print("Жаль! Всего хорошего и до свидания!")
        
    else:
        print("Ошибка! Пожалуйста, ответьте 'да' или 'нет'.")

# Запуск программы
if __name__ == "__main__":
    main()
```
