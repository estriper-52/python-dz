```python
def main():
    print("Угадаю твое число! Отвечай: '<', '>' или '='")
    print("Для выхода введи 'exit'")
    
    # Задаем регион
    try:
        left = int(input("От числа: "))
        right = int(input("До числа: "))
    except:
        print("Нужны числа!")
        return
    
    popitki = 0
    
    while left <= right:
        popitki += 1
        mid = (left + right) // 2
        
        answer = input(f"Это {mid}? ").strip().lower()
        
        if answer == 'exit':
            print("Выход")
            return
        elif answer == '=':
            print(f"Угадал за {popitki} попыток!")
            return
        elif answer == '<':
            right = mid - 1
        elif answer == '>':
            left = mid + 1
        else:
            print("Не понял, используй: < > = exit")
            popitki -= 1
    
    print("Ты где-то ошибся в ответах!")

if __name__ == "__main__":
    main()
```
