```python
# Просим пользователя ввести число
user_number = int(input("Введите число для треугольника Паскаля: "))

# Если число больше 9, будем делать только до 9, чтобы красиво получилось
if user_number > 9:
    print("Слишком большое число! Сделаю до 9 чтобы красиво было")
    user_number = 9

# Здесь будем хранить все строки треугольника
pascal_triangle = []

# Создаем каждую строку треугольника
for current_row in range(user_number):
    # Начинаем строку с единицы
    row = [1]
    
    # Для всех строк кроме первой считаем числа посередине
    if current_row > 0:
        # Берем предыдущую строку для расчетов
        prev_row = pascal_triangle[current_row - 1]
        
        # Считаем числа в середине строки
        for i in range(len(prev_row) - 1):
            # Каждое число - сумма двух чисел сверху
            new_number = prev_row[i] + prev_row[i + 1]
            row.append(new_number)
        
        # Заканчиваем строку единицей
        row.append(1)
    
    # Добавляем готовую строку в треугольник
    pascal_triangle.append(row)

# Смотрим какая самая длинная строка получилась (последняя)
last_row_str = ' '.join(str(x) for x in pascal_triangle[-1])
max_length = len(last_row_str)

# Теперь красиво печатаем весь треугольник
print("\nВот ваш треугольник Паскаля:")
for row in pascal_triangle:
    # Превращаем числа в строку с пробелами
    row_text = ' '.join(str(num) for num in row)
    # Центрируем строку чтобы был ровный треугольник
    centered_row = row_text.center(max_length)
    print(centered_row)

print("\nГотово!")  # Просто чтобы было понятно что все
```
