```python
print('izi_fibonacci')

import math

def fibonacci_number(n):
    if n == 0: return 0
    elif n == 1: return 1
    
    # Золотое сечение и его "близкий"
    phi = (1 + math.sqrt(5)) / 2
    psi = (1 - math.sqrt(5)) / 2
    
    # Считаем по формуле Бине
    numerator = power_calc(phi, n) - power_calc(psi, n)
    denominator = math.sqrt(5)
    
    return numerator / denominator

def power_calc(base, exponent):
    if exponent == 0: return 1
    # Рекурсивно умножаем
    return base * power_calc(base, exponent - 1)

# Запрос числа у пользователя
user_input = input("Введите число, не букву, позязя: ")
try:
    num = int(user_input)
    result = fibonacci_number(num)
    print(f"Для n = {num} число Фибоначчи: {result}")
except ValueError:
    print("Йоу, вообще-то я просил число, а не буквы!")
```
